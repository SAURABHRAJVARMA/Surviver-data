```python
import pandas as pd

# Load your dataset
df = pd.read_csv("/content/train.csv")

# Check for missing values
missing_data = df.isnull().sum()
print(missing_data)

# Remove rows with any missing values
df_cleaned = df.dropna()

# Remove columns with any missing values
df_cleaned = df.dropna(axis=1)

# Fill missing values with the mean of the column
df['Age'].fillna(df['Age'].mean(), inplace=True)

# Fill missing values with the median of the column
# Fill missing values with the mode of the column
df['Cabin'].fillna(df['Cabin'].mode()[0], inplace=True)

# Alternatively, fill with a constant value
# df['Cabin'].fillna('Unknown', inplace=True)

# Fill missing values with the mode of the column
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)

# Convert data types
df['Age'] = df['Age'].astype('int')

# Define lower and upper bounds
lower_bound = 10  # Example lower bound
upper_bound = 80  # Example upper bound

# Handle outliers by capping values
df['Age'] = df['Age'].clip(lower=lower_bound, upper=upper_bound)

# Convert 'Embarked' column to one-hot encoded columns
df = pd.get_dummies(df, columns=['Embarked'])
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Load your dataset
df = pd.read_csv("/content/train.csv")

# Example: Plotting a histogram of 'Age' for each 'Sex'
plt.figure(figsize=(10, 6))

# Separate data for male and female
male_age = df[df['Sex'] == 'male']['Age']
female_age = df[df['Sex'] == 'female']['Age']

# Plot histograms for male and female
plt.hist([male_age, female_age], bins=10, label=['Male', 'Female'])

# Adding titles and labels
plt.title('Age Distribution by Sex')
plt.xlabel('Age')
plt.ylabel('Frequency')

# Adding a legend
plt.legend()

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file
df = pd.read_csv('/content/train.csv')

# Filter rows where 'Pclass' equals 1 and 'Survived' equals 1
filtered_df = df[(df['Pclass'] == 1) & (df['Survived'] == 1)]

# Select 'Name', 'Pclass', 'Fare', 'Age', 'Ticket', 'Cabin', and 'Survived' columns
name_pclass_fare = filtered_df[['Name', 'Pclass', 'Fare', 'Age', 'Ticket', 'Cabin', 'Survived']]

# Plotting the Fare distribution
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)  # 1 row, 2 columns, plot 1
sns.histplot(name_pclass_fare['Fare'], kde=True, color='blue')
plt.title('Fare Distribution (Pclass 1 & Survived)')
plt.xlabel('Fare')
plt.ylabel('Count')

# Plotting the Age distribution
plt.subplot(1, 2, 2)  # 1 row, 2 columns, plot 2
sns.histplot(name_pclass_fare['Age'], kde=True, color='green')
plt.title('Age Distribution (Pclass 1 & Survived)')
plt.xlabel('Age')
plt.ylabel('Count')

# Show the plots
plt.tight_layout()
plt.show()
```
